# Analisis de Medallas por pais


```python
medallas_por_pais = df.groupby('Pais')['Total'].sum() #agrupa e total de las medallas por pais y muestra un total en conjunto
medallas_por_pais
```

```python
medallas_por_pais.sort_values(ascending=False) #Ordenando los paises con mas medallas de primero
```
